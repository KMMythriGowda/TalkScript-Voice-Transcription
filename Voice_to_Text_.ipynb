```python
def transcribe_ogg(file):
    # Convert .ogg to .wav
    audio = AudioSegment.from_ogg(file.name)
    wav_file = "temp.wav"
    audio.export(wav_file, format="wav")

    # Transcribe using AI (Google Speech Recognition)
    r = sr.Recognizer()
    with sr.AudioFile(wav_file) as source:
        audio_data = r.record(source)

    try:
        text = r.recognize_google(audio_data)
    except sr.UnknownValueError:
        text = "Sorry, could not understand the audio"
    except sr.RequestError as e:
        text = f"Could not request results; {e}"

    # Save transcript to a text file
    transcript_file = "transcript.txt"
    with open(transcript_file, "w") as f:
        f.write(text)

    return text, transcript_file
```

```python
import gradio as gr
```

```python
import speech_recognition as sr
from pydub import AudioSegment
```

```python
def transcribe_ogg(filepath):
    # Convert .ogg to .wav
    try:
        audio = AudioSegment.from_ogg(filepath)  # Use the path directly
        wav_file = "temp.wav"
        audio.export(wav_file, format="wav")
    except Exception as e:
        return f"Error converting audio: {e}", None

    # Transcribe using Google Speech Recognition
    r = sr.Recognizer()
    try:
        with sr.AudioFile(wav_file) as source:
            audio_data = r.record(source)
        text = r.recognize_google(audio_data)
    except sr.UnknownValueError:
        text = "Sorry, could not understand the audio"
    except sr.RequestError as e:
        text = f"Could not request results; {e}"

    # Save transcript to file
    transcript_file = "transcript.txt"
    with open(transcript_file, "w") as f:
        f.write(text)

    return text, transcript_file
```

```python
iface = gr.Interface(
    fn=transcribe_ogg,
    inputs=gr.Audio(type="filepath", label="Upload Voice Note (.ogg)"),
    outputs=[
        gr.Textbox(label="Transcribed Text"),
        gr.File(label="Download Transcript")
    ],
    title="TalkScript: AI-Powered Multilingual Voice Transcription System",  # <-- New App Name Here
   description="Upload any voice note (.ogg). TalkScript Converter AI will transcribe it to text instantly and let you download the transcript.")
iface.launch()
```

```python
#updated one by adding language option

import gradio as gr
import speech_recognition as sr
from pydub import AudioSegment

def transcribe_ogg(filepath, language):
    try:
        audio = AudioSegment.from_ogg(filepath)
        wav_file = "temp.wav"
        audio.export(wav_file, format="wav")
    except Exception as e:
        return f"Error converting audio: {e}", None

    r = sr.Recognizer()
    try:
        with sr.AudioFile(wav_file) as source:
            audio_data = r.record(source)
        text = r.recognize_google(audio_data, language=language)
    except sr.UnknownValueError:
        text = "Could not understand audio."
    except sr.RequestError as e:
        text = f"API error: {e}"

    transcript_file = "transcript.txt"
    with open(transcript_file, "w") as f:
        f.write(text)

    return text, transcript_file

iface = gr.Interface(
    fn=transcribe_ogg,
    inputs=[
        gr.Audio(type="filepath", label="Upload Voice Note (.ogg)"),
        gr.Dropdown(["en-US", "hi-IN"], label="Select Language", value="en-US")
    ],
    outputs=[
        gr.Textbox(label="Transcribed Text"),
        gr.File(label="Download Transcript")
    ],
    title="TalkScript: AI-Powered Voice Transcription System",
    description="Upload voice notes and get instant AI-based transcription."
)

iface.launch(share=True)
```
